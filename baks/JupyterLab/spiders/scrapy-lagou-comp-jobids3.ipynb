```python
#爬取
import scrapy
import json
import time
import math
import random




class MySpider(scrapy.Spider):
    name = 'jobs'
    rate = 1
    custom_settings = {
        'LOG_LEVEL': 'ERROR',
        'download_delay':'1'
    }
    jobCount=0 #全部工作数
    pageCount=0 #分页数
    listPages={} #全部分页
    listDone=False #分页读取完毕
    t=time.time()
    
    
    #先抓取此公司职位数量
    def start_requests(self):
        target=genTarget(0)
        req=scrapy.FormRequest(url=target.url,
                                 headers=target.headers,
                                 formdata=target.formdata,
                                 method='POST',
                                 callback=self.getCount,
                                 dont_filter=True)
        req.meta['referrer_policy']='no-referrer-when-downgrade'
        yield req
        
    #取得总数，抓取所有分页的jobid
    def getCount(self,response):
        obj=json.loads(response.text)
        if obj.get('content'):
            self.jobCount=int(obj['content']['data']['page']['totalCount'])
            self.pageCount=math.ceil(self.jobCount/10)
            print('>>Total count',self.jobCount)
            
            #填充listPages
            for i in range(1,self.pageCount+1):
                self.listPages[i]=0 #0未启动post，1请求中，2完成
            
            #循环启动每个页面的请求
            while not self.listDone:              
                #获取第一个还没启动0的页码
                pageN=0
                doneN=0
                for key in self.listPages:
                    if self.listPages[key]<1:
                        if pageN==0:
                            pageN=int(key)
                    elif self.listPages[key]==2:
                        doneN+=1                       
                        
                if doneN>=self.pageCount:
                    self.listDone=True
                    
                #启动请求,休眠0.1秒后启动
                req=None
                if pageN!=0:
                    target=genTarget(pageN)                    
                    self.listPages[pageN]=1 
                    time.sleep(1)  
                    print('>>>>>>>>>>',int(time.time())-self.t)
                    req=scrapy.FormRequest(url=target.url,
                                         headers=target.headers,
                                         formdata=target.formdata,
                                         method='POST',
                                         callback=self.getJobIds,
                                         dont_filter=True)
                    req.meta['listPageN']=pageN
                    req.meta['referrer_policy']='no-referrer-when-downgrade'
#                     print('>>>>>>>',pageN,self.listPages)
#                     print('XXXX',req.headers)
                    self.listPages[pageN]=1
                yield req
            print('>>>>>>>>>>>>>OK!!!')
        
    #取得jobid列表，
    def getJobIds(self,response):
        pageN=int(response.meta['listPageN'])
        print('>>getJobIds：listPageN',response.meta['listPageN'],int(time.time())-self.t)
        obj=json.loads(response.text)
        if obj.get('content'):
            print('--:OK',response.meta['listPageN'])
            self.listPages[pageN]=2
        else:
            self.listPages[pageN]=0
            print('x-:ERR',response.meta['listPageN'])
        print(response.text[0:50])
        
    def parse(self,response):
        pass
        
        
        
        
        
```
